```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Simulate Data

N = 100
D = 100

X = np.random.randn(N,D)

PHI = np.column_stack((np.array([1]*N), X))

w_true = np.array([3.8, 5.3, 3.9]+[0]*(D-2))

y = PHI.dot(w_true) + np.random.randn(N)
```

```python
#Functions

def objective(N, y, y_hat,lmda, w):
    return (1/2*N)*np.sum((y-y_hat)**2)+ (lmda/N)*np.sum(np.abs(w))
```

```python
w = np.random.randn(D+1)
J = []
lmda = 1000
eta = 1e-4
epochs = 1000

for t in range(epochs):
    y_hat = PHI.dot(w)
    J.append(objective(N, y, y_hat,lmda, w))
    w -= eta * (1/N) * (PHI.T.dot(y_hat - y) + lmda*np.sign(w))
    
plt.figure(figsize = (12,9))
plt.plot(J)
```
