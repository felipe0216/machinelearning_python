```python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
%matplotlib inline
```

In the event that we have a very large number of features, Gradient Descent will be faster than the normal Equation Solution.

Also, in the event that our matrix, $\PHI.T PHI\$ is singular. The nromal Equation Solution will not exist. However, Gradient Descent will still allow us to optimse our model. 

```python
#Simulate Data

N = 100

x = np.linspace(0, 10 , N)

PHI = np.vstack((np.array([1]*N),x)).T

w_true = np.array([3.9,9.9])

y = PHI.dot(w_true) + np.random.randn (N)*3 

plt.figure (figsize = (12 , 9))

plt.scatter(x , y)
```

```python
#Functions

def OLS (y, y_hat):
    return (1/(2*N))*(y - y_hat).dot(y-y_hat)
```

```python
#Fit the model with Gradient Descent

w = np.random.randn(2)

J = []
eta = 3e-2
epochs = int(1e3)

for t in range(epochs):
    y_hat = PHI.dot(w)
    J.append(OLS(y, y_hat))
    w -= eta * (1/N) * PHI.T.dot(y_hat - y)
    

plt.figure (figsize = (12,9))
plt.plot(J)
```

```python
#Plot modelfit, visualize model fit

plt.figure(figsize = (12,9))
plt.scatter (x,y)
plt.plot(x, y_hat, color = "black", linewidth = 2.5)
```

```python
#Report R-squared

R2 = 1 - np.sum((y-y_hat)**2) / np.sum ((y - y.mean())**2)

print ("R-squared: {}".format(R2))
```
